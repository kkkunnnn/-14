```python
#第七題
import pandas as pd
url = "https://drive.google.com/uc?export=download&id=13YJURg236M6tqObCJykLZOStNdEqjfry"
df = pd.read_excel(url, sheet_name='Data', engine='openpyxl')
df.rename(columns={'Series_Title': '電影片名', 'Runtime': '影片長度'}, inplace=True)
df_drop = df.dropna(axis=1)
def assign_rating(rating):
    if rating >= 9.0:
        return 'A'
    elif rating < 9.0 and rating >= 8.5:
        return 'B'
    elif rating < 8.5 and rating >= 8.0:
        return 'C'
    else:
        return 'D'
df_drop['評等'] = df_drop['IMDB_Rating'].apply(assign_rating)
df_drop.set_index('電影片名', inplace=True)
print("所有列名：", df_drop.columns)
selected_columns = ['影片長度', 'IMDB_Rating', 'No_of_Votes']
if 'Meta_score' in df_drop.columns:
    selected_columns.append('Meta_score')
if 'Gross' in df_drop.columns:
    selected_columns.append('Gross')
imdb_numeric = df_drop[selected_columns]
print(imdb_numeric.head())
imdb_numeric.to_excel('imdb_numeric.xlsx', sheet_name='Data')
```

```python
#第八題
import pandas as pd
imdb_numeric = pd.read_excel('imdb_numeric.xlsx', sheet_name='Data')
imdb_numeric = imdb_numeric.apply(pd.to_numeric, errors='coerce')
imdb_numeric = imdb_numeric.dropna()
mean_values = imdb_numeric.mean()
std_values = imdb_numeric.std()
max_values = imdb_numeric.max()
min_values = imdb_numeric.min()
covariance_matrix = imdb_numeric.cov()
correlation_matrix = imdb_numeric.corr()
print("平均值：")
print(mean_values)
print("\n標準差：")
print(std_values)
print("\n最大值：")
print(max_values)
print("\n最小值：")
print(min_values)
print("\n共變異數矩陣：")
print(covariance_matrix)
print("\n相關係數矩陣：")
print(correlation_matrix)

```

```python
#第十二題
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 imdb_drop 資料集
imdb_drop = pd.read_excel('imdb_drop.xlsx', sheet_name='Data')

# 計算 'Released_Year' 項目的次數分配
year_counts = imdb_drop['Released_Year'].value_counts()

# 繪製圓餅圖
plt.figure(figsize=(8, 6))
year_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('電影上映年份圓餅圖')
plt.axis('equal')  # 使圓餅圖比例相等，讓它成為一個圓形
plt.ylabel('')  # 移除 y 軸標籤
plt.show()
```

```python
#第十四題
import pandas as pd
import matplotlib.pyplot as plt
imdb_U = pd.read_excel('imdb_U.xlsx', sheet_name='Data')
imdb_ratings = imdb_U['IMDB_Rating'].values
stems = [int(str(score)[0]) for score in imdb_ratings]
leaves = [int(str(score)[2]) for score in imdb_ratings]
plt.figure(figsize=(10, 6))
plt.stem(stems, leaves, linefmt='C0-', markerfmt='C0o', basefmt='k-')
plt.xlabel('IMDB_Rating 莖')
plt.ylabel('IMDB_Rating 葉')
plt.title('IMDB_Rating 莖葉圖')
plt.show()

```
